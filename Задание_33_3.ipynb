```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-2*np.pi, 3*np.pi, 100)
y1 = 1 * np.cos(x - 3) + 2
plt.polar(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0., 2., 1./180.)*np.pi 
plt.polar(x, [10]*len(x)) 
plt.show()
```
