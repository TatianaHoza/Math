```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-2*np.pi, 3*np.pi, 100)

y1 = 1 * np.cos(x - 3) + 2
y2 = 2 * np.cos(x - 2) + 3
y3 = 3 * np.cos(x - 1) + 1
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
