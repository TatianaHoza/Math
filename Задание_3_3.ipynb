```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

x = []
x2 = []
y = []
y2 = []

for i in range(200):
    r = 100
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
    
plt.figure(figsize = (4,4))  
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x2, y2)
plt.plot(x2, y)
plt.show()
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

x = []
x2 = []
y = []
y2 = []

for i in range(200):
    a = 200
    b = 100
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(-a ** 2 + (a ** 2 * (i ** 2 / b ** 2))))
    y2.append(-np.sqrt(-a ** 2 + (a ** 2 * (i ** 2 / b ** 2))))
    
  
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x2, y2)
plt.plot(x2, y)
plt.show()
```
