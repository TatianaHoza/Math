```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-100, 100, 1)
Y = np.arange(-100, 100, 1)
X, Y = np.meshgrid(X, Y)
Z2 = 4 - X - 3*Y
Z1 = -4-(X+50) - 3*(Y+80)
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-100, 100, 2)
Y = np.arange(-100, 100, 2)
A=30
B=10
X, Y = np.meshgrid(X, Y)
Z2 = np.sqrt((X**2/B**2) - (Y**2/A**2))
Z1 = np.sqrt(((X**2)/(A**2))-((Y**2)/(B**2)))
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
