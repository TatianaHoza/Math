```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
n = 100
r = 0.9
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
c = np.corrcoef(x, y)
k = (np.sum((x-x*r))*(y-y*r))/(np.sqrt((np.sum((x-x*r)**2))*(np.sum((y-y*r)**2))))
print(c)
print(k)
```
